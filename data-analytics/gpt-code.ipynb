```python
import pandas as pd
import seaborn as sns

# Creating a DataFrame for the medals data
data = {
    'Year': years,
    'Gold': gold,
    'Silver': silver,
    'Bronze': bronze
}
medals_df = pd.DataFrame(data)

# Creating a stacked area plot using Seaborn
plt.figure(figsize=(14, 8))
plt.stackplot(medals_df['Year'], medals_df['Gold'], medals_df['Silver'], medals_df['Bronze'], 
              labels=['Gold', 'Silver', 'Bronze'], colors=['gold', 'silver', 'brown'])

# Adding titles and labels
plt.title('US Olympic Medals Breakdown (1924 - 2024)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Medals', fontsize=14)
plt.xticks(years, rotation=45)
plt.legend(loc='upper left')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculating cumulative sums for the stacked effect
medals_df['Cumulative Gold'] = medals_df['Gold']
medals_df['Cumulative Silver'] = medals_df['Cumulative Gold'] + medals_df['Silver']
medals_df['Cumulative Bronze'] = medals_df['Cumulative Silver'] + medals_df['Bronze']

# Using Seaborn to create the plot
plt.figure(figsize=(14, 8))
sns.lineplot(data=medals_df, x='Year', y='Cumulative Bronze', label='Bronze', color='brown', lw=2.5)
sns.lineplot(data=medals_df, x='Year', y='Cumulative Silver', label='Silver', color='silver', lw=2.5)
sns.lineplot(data=medals_df, x='Year', y='Cumulative Gold', label='Gold', color='gold', lw=2.5)

plt.title('US Olympic Medals Breakdown (1924 - 2024)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Cumulative Number of Medals', fontsize=14)
plt.xticks(years, rotation=45)
plt.legend(title='Medal Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```
